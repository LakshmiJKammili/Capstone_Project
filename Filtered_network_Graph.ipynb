```python
import re
import pandas as pd

def extract_metrics(log_text):
  """Extracts the train and validation metrics from the log text.

  Args:
    log_text: The log text.

  Returns:
    A dictionary containing the train and validation metrics.
  """

  # Define regular expressions
  patterns = [
      (r'step (\d+)', 'Step'),
      (r'learning_rate = ([\d.e-]+)', 'Learning Rate'),
      (r'loss_average = ([\d.e+]+)', 'Loss Average'),
      (r'train_set accuracy: ([\d.]+)', 'Train Accuracy'),
      (r'f1 \(weighted\): ([\d.]+)', 'Train F1 (Weighted)'),
      (r'loss: ([\d.e+]+)', 'Train Loss'),
      (r'validation accuracy: ([\d.]+)', 'Validation Accuracy'),
      (r'f1 \(weighted\): ([\d.]+)', 'Validation F1 (Weighted)'),
      (r'loss: ([\d.e+]+)', 'Validation Loss')
  ]

  # Initialize data dictionary
  data = {label: [] for _, label in patterns}

  # Extract information using regular expressions
  for pattern, label in patterns:
    matches = re.findall(pattern, log_text)
    if label == 'Train F1 (Weighted)' or label == 'Train Loss':
      # Delete the second value
      matches.pop(-1)
    elif label == 'Validation F1 (Weighted)' or label == 'Validation Loss':
      # Delete the first value
      matches.pop(0)
    data[label].extend(matches)

  return data



# Extract the train and validation metrics
metrics = extract_metrics(log_text)

# Print the metrics
print(metrics)
```

```python
import pandas as pd
import plotly.graph_objects as go

# Load your CSV data
df = pd.read_csv('/content/predicted_concordance.csv')

# Assuming 'label' and 'predicted' are the column names
true_labels = df['label']
predicted_labels = df['Predicted']

# Count occurrences of each label
label_counts = true_labels.value_counts().sort_index()
predicted_counts = predicted_labels.value_counts().sort_index()

# Create an interactive bar chart
fig = go.Figure()

fig.add_trace(go.Bar(
    x=['Non-Metastasis (Actual)', 'Metastasis (Actual)'],
    y=[label_counts[0], label_counts[1]],
    name='Actual',
    marker_color='blue'
))

fig.add_trace(go.Bar(
    x=['Non-Metastasis (Predicted)', 'Metastasis (Predicted)'],
    y=[predicted_counts[0], predicted_counts[1]],
    name='Predicted',
    marker_color='orange'
))

fig.update_layout(
    title='Comparison of Actual vs. Predicted Labels',
    xaxis_title='Labels',
    yaxis_title='Counts',
    barmode='group'
)

fig.show()
```
