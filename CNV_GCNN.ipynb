```python
import pandas as pd

# Load the CSV file
file_path = '/content/smoteenn_resampled_data.csv'
df = pd.read_csv(file_path)

# Assuming 'num_synthetic_samples' is defined or calculated somewhere in your code
# For demonstration, I'll set it to a fraction of the DataFrame's length
num_synthetic_samples = len(df) // 2  # Adjust this as necessary

# Original indices (with "GSM" prefix)
original_indices = [f"GSM{i}" for i in range(len(df) - num_synthetic_samples)]

# New indices for the synthetic samples (with "GSM" prefix)
new_indices = [f"GSMnew_{i}" for i in range(num_synthetic_samples)]

# Combine the original and new indices
combined_indices = original_indices + new_indices

# Add combined_indices as a column named "sample_id"
df['sample_id'] = combined_indices

# Set "sample_id" as the index
df = df.set_index('sample_id')

# Verify the DataFrame by displaying the first few rows
df.head()
```

```python
# Split the DataFrame into SNP data and metastasis labels
cnv_data = df.drop('Metastasis', axis=1)  # Assuming 'Metastasis' is the column name to drop
metastasis_labels = df[['Metastasis']]  # Assuming 'Metastasis' is the column name to keep

# Transpose the SNP data to get genes as columns and sample IDs as rows
cnv_data_transposed = cnv_data.T

# Transpose the metastasis labels DataFrame
metastasis_labels_transposed = metastasis_labels.T

# Display the transposed metastasis labels DataFrame
print(metastasis_labels_transposed)

# Optionally, save the transposed DataFrames to new CSV files
#cnv_data_transposed.to_csv('/path/to/your/snp_data_transposed.csv')
metastasis_labels_transposed.to_csv('labels.csv')
```

```python
# Ensure the 'sample_id' column is included in the list of columns to keep, if it's not already
if 'sample_id' not in first_column_values:
    first_column_values.append('sample_id')

# Filter the DataFrame to keep only the columns in first_column_values or the 'sample_id' column
cnv_data_filtered = cnv_data.loc[:, cnv_data.columns.isin(first_column_values) | (cnv_data.columns == 'sample_id')]
```

```python
cnv_data_filtered_transposed= cnv_data_filtered.T
```

```python
# Reset the index to turn it into a regular column
cnv_data_filtered_reset = cnv_data_filtered_transposed.reset_index()

# Rename the new column to "probe"
cnv_data_filtered_reset.rename(columns={'index': 'probe'}, inplace=True)

# Get the list of all columns, move 'probe' to the end
columns = list(cnv_data_filtered_reset.columns)
columns.append(columns.pop(columns.index('probe')))  # Move 'probe' to the end

# Reorder the DataFrame using the new column order
cnv_data_filtered_reset_final = cnv_data_filtered_reset[columns]

# Display the updated DataFrame
cnv_data_filtered_reset_final
```
