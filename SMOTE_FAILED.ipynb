```python
# Re-importing pandas as the code execution state was reset
import pandas as pd

# Re-define the file paths
snp_file_path = '/content/snp_indel_data_filtered.csv'
metastasis_file_path = '/content/metastasis_filtered_corrected.csv'

# Re-load the SNP and metastasis datasets
snp_df = pd.read_csv(snp_file_path)
metastasis_df = pd.read_csv(metastasis_file_path)

# Transpose the SNP dataframe to turn rows into columns
snp_df_transposed = snp_df.T
snp_df_transposed.columns = snp_df_transposed.iloc[0]  # Set the first row as the column names
snp_df_transposed = snp_df_transposed.iloc[1:]  # Remove the first row as it's now the header

# For the metastasis dataframe, transpose it so that sample IDs become the index
metastasis_status = metastasis_df.T
metastasis_status.columns = ['Metastasis']  # Name the metastasis status column

# Join the metastasis status with the transposed SNP dataframe based on sample IDs
snp_with_metastasis = snp_df_transposed.join(metastasis_status, how='left')

# Display the last few rows to verify the structure
snp_with_metastasis.tail()
```

```python
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
import numpy as np

# Assuming 'Metastasis' is your target variable
X = snp_with_metastasis.drop('Metastasis', axis=1)  # Features
y = snp_with_metastasis['Metastasis']  # Target variable

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=42)
X_train.columns = X_train.columns.astype(str)
# Initialize SMOTE and balance the training set
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

# Check the balance
print("Before SMOTE:", np.bincount(y_train))
print("After SMOTE:", np.bincount(y_train_smote))
```

```python
from sklearn.model_selection import train_test_split
from imblearn.combine import SMOTEENN
import pandas as pd
import numpy as np

# Assuming 'Metastasis' is your target variable and 'cnv_with_metastasis' is your DataFrame
X = cnv_with_metastasis.drop('Metastasis', axis=1)  # Features
y = cnv_with_metastasis['Metastasis'].astype(int)  # Target variable

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=42)
X_train.columns = X_train.columns.astype(str)

# Display the class distribution before SMOTEENN
print("Class distribution before SMOTEENN:", np.bincount(y_train))

# Initialize SMOTEENN and balance the training set
smote_enn = SMOTEENN(random_state=42)
X_train_smoteenn, y_train_smoteenn = smote_enn.fit_resample(X_train, y_train)

# Display the class distribution after SMOTEENN
print("Class distribution after SMOTEENN:", np.bincount(y_train_smoteenn))

# Create DataFrames for the resampled data
X_train_smoteenn_df = pd.DataFrame(X_train_smoteenn, columns=X_train.columns)
y_train_smoteenn_df = pd.Series(y_train_smoteenn, name='Metastasis')

# Concatenate features and labels into one DataFrame
smoteenn_df = pd.concat([X_train_smoteenn_df, y_train_smoteenn_df], axis=1)

# Save the resampled data to CSV
smoteenn_df.to_csv('smoteenn_resampled_data.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# Load your dataset
file_path = '/content/smoteenn_resampled_data.csv'  # Adjust the file path if needed
data = pd.read_csv(file_path)

# Aggregate unique values from all columns
all_unique_values = pd.unique(data.values.ravel('K'))

# Count frequencies of each unique value
value_counts = Counter(all_unique_values)

# Prepare data for plotting
values, frequencies = zip(*value_counts.items())

# Create the plot
plt.figure(figsize=(10, 6))
plt.bar(values[:100], frequencies[:100])  # Plotting a subset for clarity
plt.title('Frequency of Unique Values Across All Columns (Subset)')
plt.xlabel('Unique Values')
plt.ylabel('Frequency')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()
```
